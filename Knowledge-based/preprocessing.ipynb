```python
import pandas as pd
import pickle
import csv
```

```python
emb = pd.read_pickle('data/books_emb.pkl')
```

```python
inter = emb[['user', 'item', 'rating']]
```

```python
inter = inter.rename(columns={
    'user': 'user_id',
    'item': 'item_id',
    'rating': 'rating'
})

with open('KGAT/dataset/amazon-book/amazon-book.inter', 'w') as f:
    f.write('user_id:token\titem_id:token\trating:float\n')
    inter.to_csv(f, sep='\t', index=False, header=False)
```

```python
item_emb = emb[['item', 'item_id']]
item_emb
```

```python
kg = pd.read_pickle('data/kg.pkl')
kg['head'] = kg['head'].str.replace('item_', '', regex=False)
kg
```

```python
item_id_map = dict(zip(item_emb['item_id'].astype(int), item_emb['item']))
kg['head'] = kg['head'].astype(int).map(item_id_map)
kg
```

```python
with open('KGAT/dataset/amazon-book/amazon-book.kg', 'w') as f:
    f.write('head_id:token\trelation_id:token\ttail_id:token\n')
    kg.to_csv(f, sep='\t', index=False, header=False)
```

```python
link = emb[['item']]
unique_item = link['item'].drop_duplicates()

link_df = pd.DataFrame({
    'item_id': unique_item,
    'entity_id': unique_item
})

link_df
```

```python
with open('KGAT/dataset/amazon-book/amazon-book.link', 'w') as f:
    f.write('item_id:token\tentity_id:token\n')
    link_df.to_csv(f, sep='\t', index=False, header=False)
```
